```python
import snowflake.connector as sf
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import snowflake.connector as sf
from scipy.stats import mannwhitneyu
from scipy.stats import ttest_ind
```

```python
# This function takes dataframe and do mannwhitneyu test and get teststats and pvalue 
def whitney_result_wpc(df):
    
    test_stats,pval=mannwhitneyu(x=df['USER_WPC'], y=df['ALPHA_RANDOM_WPC'],alternative='greater')
    return test_stats,pval
  

# This function takes dataframe and do ttest and get teststats and pvalue 

def ttest_result(df):
    
        t_stat, p_value = ttest_ind(df['USER_WPC'], df['ALPHA_RANDOM_WPC'], equal_var=False,alternative='greater')
        return t_stat, p_value
         # Set equal_var=False if variances are unequal
```

```python
#In below code we are loading the dataframes and storing in dictionary with key as alpha parameter

dict_dfs = {}  # Dictionary to store DataFrames
df=pd.read_csv('skill_gradient_data.csv', index_col=0)
for i in np.arange(0, 1.1, 0.1):
    alpha = np.round(i, 2)
    dict_dfs[alpha] = df[df['alpha_param']==alpha]
```

```python
dict_dfs
```

```python
d={}
d['alpha_parameter']=[]
d['mannwhitney_stats']=[]
d['pvalue_mannwhitneyu']=[]
d['mannwhitneyu_result']=[]
d['ttest_stats']=[]
d['ttest_pvalue']=[]
d['ttest_result']=[]
d['mean_diff_wpc']=[]
d['median_diff_wpc']=[]

for i in np.arange(0,1.1,0.1):
    
        alpha = np.round(i, 2)
        d['alpha_parameter'].append(alpha)
        df=dict_dfs[alpha] 
        res,pval = whitney_result_wpc(df)
        d['mannwhitney_stats'].append(res)
        d['pvalue_mannwhitneyu'].append(pval)

        if (pval<0.05):
            d['mannwhitneyu_result'].append('rejected')
        else:
            d['mannwhitneyu_result'].append('fail to reject')

        print(alpha,'man_whitneypvalue = ',pval)





        t_stat,p_value=ttest_result(df)
      

        print(f"T-statistic: {t_stat}, P-value: {p_value}")

        d['ttest_stats'].append(t_stat)
        d['ttest_pvalue'].append(p_value)

        if p_value<=0.05:
            
            #considering confidence level of 95%
            
            d['ttest_result'].append('significant')
        else:
            d['ttest_result'].append('insignificant')


        print(alpha,'ttest_pvalue = ',pval)

        mean=np.mean(np.abs(df['USER_WPC']- df['ALPHA_RANDOM_WPC']))
        median=np.median(np.abs(df['USER_WPC']- df['ALPHA_RANDOM_WPC']))
      
        d['mean_diff_wpc'].append(mean)
        d['median_diff_wpc'].append(median)

test_output=pd.DataFrame(d)
test_output
```
