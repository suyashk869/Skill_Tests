```pyspark
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
from scipy.stats import jarque_bera
from scipy.stats import shapiro
from scipy.stats import spearmanr
from datetime import datetime
import snowflake.connector as sf
```

```pyspark
#load data and append the monthly data making leading month_no as key
dc={}
data=pd.read_csv("skill_persistence_data_2024.csv", index_col=0)
for m1 in range(2,13):
   
    dc[m1]=data[data['MONTH_1']==m1]
    
    
```

```pyspark
d={}
d['month']=[]
d['year']=[]
d['statistic']=[]
d['pvalue']=[]
d['result']=[]




year=2024

for m1 in range(2,13):
    
    d['month'].append(m1)
    d['year'].append(year)
    
    df=dc[m1]
    df=df.astype('float64')
    
    # getting monthwise spearman correlation coefficient
    
    statistic = spearmanr(df['WIN_RATE_2'],df['WIN_RATE_1'],alternative='greater')[0]
    pvalue = spearmanr(df['WIN_RATE_2'],df['WIN_RATE_1'],alternative='greater')[1]
    if pvalue<0.05:
        result='significant'
    else :
        result='insignificant'
                
                            
        
            
    d['statistic'].append(statistic)
    d['pvalue'].append(pvalue)

    d['result'].append(result)

                
               
                
output=pd.DataFrame(d)  
output
```
